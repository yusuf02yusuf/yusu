```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://editorial.rottentomatoes.com/guide/dc-movies-in-order/"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
name=soup.select(".article_movie_title a")
```

```python
name
```

```python
year=soup.select(".start-year")
```

```python
year
```

```python
rating=soup.select(".span-icon-audience+ .tMeterScore , #row-index-11 .start-year")
```

```python
rating
```

```python
import pandas as pd
```

```python
movielist=pd.DataFrame(name,columns=['movie title'])
movielist["year"]=year
movielist["rating"]=rating
```

```python
movielist
```

```python
movielist.to_csv('D:/yusuf/dcmovies.csv')
```
